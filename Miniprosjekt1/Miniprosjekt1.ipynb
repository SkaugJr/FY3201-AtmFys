```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
#Data hentet fra seklima.met.no

df = pd.read_csv('TrdVoll.csv', usecols=[3, 4, 5, 6], skiprows=1, delimiter=';')

df.columns = ['MidTemp', 'MaxTemp', 'MinTemp', 'Nedbør']

df['MidTemp'] = df['MidTemp'].str.replace(',', '.').replace('-', np.nan).astype(float)
df['MaxTemp'] = df['MaxTemp'].str.replace(',', '.').replace('-', np.nan).astype(float)
df['MinTemp'] = df['MinTemp'].str.replace(',', '.').replace('-', np.nan).astype(float)
df['Nedbør'] = df['Nedbør'].str.replace(',', '.').replace('-', np.nan).astype(float)

MidTempTrd = df['MidTemp'].to_numpy()
MaxTempTrd = df['MaxTemp'].to_numpy()
MinTempTrd = df['MinTemp'].to_numpy()
NedbørTrd = df['Nedbør'].to_numpy()

DagDiff = MaxTempTrd - MinTempTrd

start_date = datetime.datetime(1997, 1, 1)
time_axis = pd.date_range(start=start_date, periods=len(MidTempTrd), freq='D')
```

```python
plt.figure(figsize=(15, 18))


plt.subplot(5, 1, 1)
plt.plot(time_axis, MidTempTrd, label='MidTempTrd')
plt.xlabel('År')
plt.ylabel('Temperatur [C]')
plt.title('MidTempTrd Over Tid')
plt.legend()
plt.grid(True)


plt.subplot(5, 1, 2)
plt.plot(time_axis, MaxTempTrd, label='MaxTempTrd')
plt.plot(time_axis, MinTempTrd, label='MinTempTrd')
plt.xlabel('År')
plt.ylabel('Temperatur [C]')
plt.title('MaxTempTrd and MinTempTrd Over Tid')
plt.legend()
plt.grid(True)


plt.subplot(5, 1, 3)
plt.plot(time_axis, DagDiff, label='DagDiff')
plt.xlabel('År')
plt.ylabel('Temperatur [C]')
plt.title('Differanse mellom Max og Min Temperatur Over Tid')
plt.legend()
plt.grid(True)


plt.subplot(5, 1, 4)
plt.plot(time_axis, NedbørTrd, label='NedbørTrd')
plt.xlabel('År')
plt.ylabel('Nedbør [mm]')
plt.title('NedbørTrd Over Tid')
plt.legend()
plt.grid(True)


plt.subplot(5, 1, 5)
plt.hist(NedbørTrd, bins=32, edgecolor='black')
plt.xlabel('Nedbør [mm]')
plt.ylabel('Frekvens')
plt.title('Histogram av NedbørTrd')
plt.xlim(left=0,right=30)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
